```python
kabul = df[df['location_name'] == 'Kabul'].copy()
kabul['last_updated']= pd.to_datetime(kabul['last_updated'])
kabul = kabul.sort_values('last_updated').reset_index(drop = True)
split_point = int(len(kabul)*0.8)
train = kabul.iloc[:split_point]
test = kabul.iloc[split_point:]
print(train.shape,test.shape)
```

```python
train = train.copy()
test = test.copy()
train['days_since_start'] = (train['last_updated'] - kabul['last_updated'].min()).dt.days
test['days_since_start'] = (test['last_updated'] - kabul['last_updated'].min()).dt.days
train[['last_updated','days_since_start']].head()
```

```python
from sklearn.linear_model import LinearRegression
X_train = train[['days_since_start']]
y_train = train['temperature_celsius']

X_test = test[['days_since_start']]
y_test = test['temperature_celsius']

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
predictions[:5]
```

```python
plt.figure(figsize=(12,5))
plt.plot(test['last_updated'], y_test,label = "Actual",marker ='o')
plt.plot(test['last_updated'], predictions,label = "Predicted",marker ='x')
plt.title("Actual vs Predicted Temperature - Kabul (Test Set)")
plt.xlabel("Date")
plt.ylabel("Temperature(°C)")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import numpy as np
mae = mean_absolute_error(y_test,predictions)
rmse = np.sqrt(mean_squared_error(y_test,predictions))
r2 = r2_score(y_test,predictions)
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R**2: {r2:.2f}")
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
model_poly = LinearRegression()
model_poly.fit(X_train_poly,y_train)
predictions_poly = model_poly.predict(X_test_poly)
```

```python
mae_poly = mean_absolute_error(y_test,predictions_poly)
rmse_poly = np.sqrt(mean_squared_error(y_test,predictions_poly))
r2_poly = r2_score(y_test,predictions_poly)
print(f"MAE: {mae_poly:.2f}")
print(f"RMSE: {rmse_poly:.2f}")
print(f"R**2: {r2_poly:.2f}")
```
