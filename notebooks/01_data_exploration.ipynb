```python
location_counts = df['location_name'].value_counts()
location_counts
```

```python
kabul = df[df['location_name'] == 'Kabul']
```

```python
kabul['last_updated'] = pd.to_datetime(kabul['last_updated'])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,7))
```
